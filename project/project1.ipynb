```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/sample_data/Dec 2024.csv")
df
```

```python
region_mapping = {
    1: "East",
    2: "North",
    3: "South",
    4: "West"
}
df["Region"] = df["Region ID"].map(region_mapping)
```

```python
sales_by_region = df.groupby("Region")["Revenue"].sum().sort_index()
sales_by_region_millions = sales_by_region / 1_000_000
```

```python
plt.figure(figsize=(6, 5))
bars = plt.bar(sales_by_region_millions.index, sales_by_region_millions.values, color="blue")
```

```python
plt.title("Total Sales By Region", fontsize=14)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Sales (in Millions)", fontsize=12)
plt.ylim(0, sales_by_region_millions.max() + 2)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
